```python
import numpy as np
import matplotlib.pyplot as plt
```

$ h(r) = a_1 x^3 + a_2 x^2 + a_3 x + a_0 $

such that 

* $h(0)=0$
* $h(r_+) = h_\infty$
* $h'(0) = h'_0$ 
* $h''(x) < 0$
* $h''(r_+) = 0 $

doing some calculations and assuming $a_0=0$ results in

* $a_1 = \frac{-1}{r^3_+}(2h_\infty + r_+ h'_0)$
* $a_2 = \frac{1}{r_+} \left( \frac{3}{r_+}h_\infty - 2 h'_0\right)$
* $a_3 = h'_0$

```python
h0=0.1
hinf=300000
rbar=1
```

```python
a1 = -1 / rbar / rbar / rbar * (2*hinf + rbar * h0)
a2 = 1 / rbar * ( (3/rbar) * hinf - 2 * h0 )
a3 = h0
```

```python
def h(r):
    return a1 * r * r * r + a2 * r * r + a3 *r
```

```python
x = np.linspace(0,1.5)
```

```python
plt.plot(x,h(x))
plt.grid()
plt.ylabel(r"h(r)")
plt.xlabel(r"r")
```

```python
def hp(r):
    return 3 *a1 * r * r + 2 * a2 * r + a3
```

```python
x = np.linspace(0,rbar)
plt.plot(x,hp(x))
plt.grid()
plt.ylabel(r"h'(r)")
plt.xlabel(r"r")
```

```python
def hpp(r):
    return 6*a1*r+2*a2
```

```python
x = np.linspace(0,rbar)
plt.plot(x,hpp(x))
plt.grid()
plt.ylabel(r"h''(r)")
plt.xlabel(r"r")
```

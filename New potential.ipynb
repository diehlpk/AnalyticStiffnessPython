```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 3
f = 1
r = 1
```

 $g(x) = \begin{cases} (x-1)^{2n},  0 \leq x < 1 \\
  0 , x \geq 0\end{cases}$

```python
def g(x,n):
    return np.power((np.power(x,2)/r-1),2*n)
    
```

```python
x = np.linspace(0,1)
```

```python
plt.plot(x,g(x,n))
plt.title(r"$g(x)$")
plt.grid()
```

$f(x^2) = f_\infty \left( 1 - g(\frac{r^2}{r_+})\right)$ 

```python
x = np.linspace(0,r)
```

```python
plt.plot(x,f*(1-g(x,n)),label=r"$\Psi(x)$")
plt.grid()
plt.axvline(x=np.sqrt(r/(4*n-1)),label=r"$r_c=\sqrt{\frac{r_+}{2n}}$",c="black")
plt.legend()
```

$\Psi'(x) = - \frac{1}{r}4 f_\infty n x \left( \frac{x^2}{r}-1 \right)^{2*n-1}$

```python
def fp(x):
    return -4*f * n * x * np.power(x*x/r-1,2*n-1) / r
```

```python
plt.plot(x,fp(x),label=r"$\Psi'(x)$")
plt.axvline(x=np.sqrt(r/(4*n-1)),label=r"$r_c=\sqrt{\frac{r_+}{2n}}$",c="black")
plt.grid()
plt.legend()
```

$r_c=\sqrt{\frac{r_+}{2n}}$

$ \Psi''(x) -\frac{4 f n {{\left( \frac{{{x}^{2}}}{r}-1\right) }^{2 n-1}}}{r}-\frac{8 f n\, \left( 2 n-1\right)  {{x}^{2}} {{\left( \frac{{{x}^{2}}}{r}-1\right) }^{2 n-2}}}{{{r}^{2}}}$

```python
def fpp(x):
    return -(4*f*n*np.power((x*x/r-1),(2*n-1)))/r-(8*f*n*(2*n-1)*x*x*np.power((x*x/r-1),(2*n-2)))/r/r
```

```python
plt.plot(x,fpp(x),label=r"$\Psi''(x)$")
plt.axvline(x=np.sqrt(r/(4*n-1)),label=r"$r_c=\sqrt{\frac{r_+}{2n}}$",c="black")
plt.grid()
plt.legend()
```
